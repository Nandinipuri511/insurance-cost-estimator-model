LINEAR REGRESSION

```python
non_smoker = medical_df[medical_df.smoker == 0]
```

```python
plt.title ("Age wrt charges")
sns.scatterplot(data = non_smoker , x = 'age', y = 'charges', alpha = 0.7,s = 15)
```

LINEAR REGRESSION MODEL

charges = w * age + b *italicised text*
here w -> weight and b -> bias
in mathematical terms w -> slope, b -> intercept

values in age col of dataset are called **inputs** and values in charges cols are called **targets**

This is known as linear regression model since it models the relationship bw age and charges as a straight line

```python
def estimate_charges(age, w, b):
  return w * age + b
```

```python
w = 50
b = 100
```

```python
estimate_charges(45, w, b)
```

```python
ages = non_smoker.age
ages
```

```python
est_charges = estimate_charges(ages, w, b)
est_charges
```

```python
non_smoker.charges
```

```python
plt.plot(ages,est_charges, 'm-p')
plt.xlabel('Age')
plt.ylabel('Estimated Charges')
```

```python
target = non_smoker.charges

plt.plot(ages, est_charges, 'r', alpha = 0.9);
plt.scatter(ages, target, s = 8, alpha = 0.7);
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend(['Estimate', 'Actual'])
```

From the above graph it is clearly demonstrated that the line doesn't **fit** the data...
so we'll try using different values of w and b to fit the line and for doing this we'll make use of a fuction try_parametrs which takes w and b as inputs and creates the above plot.

```python
def try_parameters(w, b):
  ages = non_smoker.age
  target = non_smoker.charges

  est_charges = estimate_charges(ages, w, b)

  plt.plot(ages, est_charges, 'r', alpha = 0.9);
  plt.scatter(ages, target, s = 8, alpha = 0.7);
  plt.xlabel('Age')
  plt.ylabel('Charges')
  plt.legend(['Estimate', 'Actual'])

  loss = rmse(target, est_charges)
  print(f'Loss: {loss}')
```

```python
try_parameters(60,200)
```

```python
try_parameters(120,500)
```

LOSS/COST FUNCTION


*   Calc diff bw targets and predictions(diff is aka residual)
*   Square all elements of diff matrix to remove negative value
*   Calculate avg of elements in resulting matrix
*   Take the square root of the result

the result is single number called RMSE(root mean squared error)





```python
targets = non_smoker.charges
targets
```

```python
predictions = estimate_charges(ages, w, b)
predictions
```

```python
import numpy as np
def rmse(targets, predictions):
  return np.sqrt(np.mean(np.square(targets-predictions)))
```
